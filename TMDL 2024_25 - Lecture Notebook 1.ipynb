Let's look at genres now. We first should process the strings in the genre column to a list of useful genre names (we don't really care about the ids for now).

```python
df["genres"].iloc[0]
```

```python
import json

def process_genre_string(gs):
    gs = json.loads(gs)
    gs = [x["name"] for x in gs]
    return gs
```

```python
process_genre_string(df["genres"].iloc[0])
```

```python
df["genres"].map(process_genre_string)
```

```python
df["proc_genres"] = df["genres"].map(process_genre_string)
df[["proc_genres", "title"]]
```

In machine learning, the concepts of overfitting and underfitting are crucial for understanding model performance.

- **Underfitting:** This occurs when a model is too simple to capture the underlying patterns in the data, leading to poor training and testing performance.
- **Overfitting:** This happens when a model is too complex and captures noise in the training data, leading to high training accuracy but poor generalization to new data.
- **Ideal Model Fit:** A well-fit model achieves a balance between bias (underfitting) and variance (overfitting), leading to good generalization.

To diagnose overfitting or underfitting, we can examine learning curves by plotting training and validation loss over iterations.


```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated training and validation loss
epochs = np.arange(1, 21)
overfitting_train_loss = np.exp(-0.3 * epochs) + 0.1 * np.random.randn(20)
overfitting_val_loss = np.exp(-0.3 * epochs) + 0.4 * np.random.randn(20) + 0.3

underfitting_train_loss = np.linspace(1, 0.7, 20) + 0.1 * np.random.randn(20)
underfitting_val_loss = np.linspace(1.2, 1.0, 20) + 0.2 * np.random.randn(20)

plt.figure(figsize=(10, 5))

# Overfitting plot
plt.subplot(1, 2, 1)
plt.plot(epochs, overfitting_train_loss, label='Training Loss')
plt.plot(epochs, overfitting_val_loss, label='Validation Loss')
plt.title("Overfitting: Large gap between training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

# Underfitting plot
plt.subplot(1, 2, 2)
plt.plot(epochs, underfitting_train_loss, label='Training Loss')
plt.plot(epochs, underfitting_val_loss, label='Validation Loss')
plt.title("Underfitting: High loss in both training and validation")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.show()
```

### Graph Interpretation: Loss vs. Model Fit
- If training and validation loss decrease and stabilize at a similar level, the model is well-fit.
- If validation loss starts increasing while training loss keeps decreasing, the model is overfitting.
- If both losses remain high, the model is underfitting.

By examining these patterns, we can adjust the model complexity, increase training data, or use regularization techniques to improve performance.
